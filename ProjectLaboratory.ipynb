<h3>This is inference stage, in which you can generate video with prompt. Pay attention to use correct pathes for files. Also, according to the number of epochs carried out during training, the name of "model" file changes. For example, in the code below it is "ddim_latent-200.pt", because the model has been executed for 200 epochs.

```python
from main.tuneavideo.pipelines.pipeline_tuneavideo import TuneAVideoPipeline
from main.tuneavideo.models.unet import UNet3DConditionModel
from main.tuneavideo.util import save_videos_grid
import torch

pretrained_model_path = "/content/stable-diffusion-v1-4"
my_model_path = "/content/drive/MyDrive/ProjectLaboratory/car_turn"
unet = UNet3DConditionModel.from_pretrained(my_model_path, subfolder='unet', torch_dtype=torch.float16).to('cuda')
pipe = TuneAVideoPipeline.from_pretrained(pretrained_model_path, unet=unet, torch_dtype=torch.float16).to("cuda")
pipe.enable_xformers_memory_efficient_attention()
pipe.enable_vae_slicing()

prompt = "a tank is moving in New-York"
ddim_inv_latent = torch.load(f"{my_model_path}/inv_latents/ddim_latent-200.pt").to(torch.float16)
video = pipe(prompt, latents=ddim_inv_latent, video_length=24, height=512, width=512, num_inference_steps=50, guidance_scale=12.5).videos

save_videos_grid(video, f"./{prompt}.gif")
```

<h2>After running this code, you will get a .gif file with the file name same as prompt.
